```python
def predd(x, S1, S2, S3, S4, S5, S6, S7, S8):
    psymptoms = [S1, S2, S3, S4, S5, S6, S7, S8]

    # Convert symptoms to weights
    a = np.array(df["Symptom"])
    b = np.array(df["weight"])

    for j in range(len(psymptoms)):
        for k in range(len(a)):
            if psymptoms[j] == a[k]:
                psymptoms[j] = b[k]
    
    # Replace unrecognized symptoms with 0 (optional safety check)
    for i in range(len(psymptoms)):
        if isinstance(psymptoms[i], str):  # If not converted to a weight
            psymptoms[i] = 0

    # Predict the disease
    psy = [psymptoms]
    pred2 = x.predict(psy)
    predicted_disease = pred2[0]

    # Disease description
    disp = disc[disc['Disease'] == predicted_disease]
    if not disp.empty:
        description = disp.values[0][1]
    else:
        description = "No description available."

    # Precautions
    precaution_list = []
    if predicted_disease in precaut['Disease'].values:
        c = np.where(precaut['Disease'] == predicted_disease)[0][0]
        for i in range(1, len(precaut.iloc[c])):
            precaution = precaut.iloc[c, i]
            if isinstance(precaution, str) and precaution.strip() != "":
                precaution_list.append(precaution)

    # Output results
    print("\n🦠 Predicted Disease:", predicted_disease)
    print("📄 Description:", description)
    print("✅ Precautions:")
    for p in precaution_list:
        print("-", p)

    return predicted_disease  # Return so it can be used in the doctor recommendation
```
